```python
## Imports
import numpy as np
import pandas as pd
import requests
import json

# ML
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# PyTorch 
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

from tqdm import tqdm

import tensorflow as tf
from tensorflow.keras import layers, models

import matplotlib.pyplot as plt
```

Pull Fluorescent Protein Database

```python
url = "https://www.fpbase.org/api/proteins/?format=json"

response = requests.get(url)
data = response.json()

print(type(data))
print(len(data))
```

Transform content of Db to Df

```python
rows = []

for protein in data:
    
    states = protein.get("states", [])
    if len(states) == 0:
        continue
    
    state = states[0]
    
    rows.append({
        "name": protein.get("name"),
        "sequence": protein.get("seq"),
        "aggregation_type": protein.get("agg"),
        "excitation": state.get("ex_max"),
        "emission": state.get("em_max"),
        "brightness": state.get("brightness") 
    })

df = pd.DataFrame(rows)
df.head()
```

Drop lines with lacking values for the parameters needed later

```python
df_no_empty_lines = df.dropna(subset=["sequence", "emission", "emission","brightness"])
df_no_empty_lines.shape
df_no_empty_lines.isna().sum()

```

Since sequences may contain peptide tags, they are replaced by tag-free sequences from the Protein Data Base (PDB)

```python
def get_pdb_sequence(name):
    """
    Obtain PDB sequence of given protein.
    Returns PDB sequence if protein in PDB, else None.
    """
    url = f"https://www.rcsb.org/fasta/entry/{name}"
    try:
        response = requests.get(url)
        if response.status_code == 200:
            #FASTA format: first line is header, second line is sequence, so ignore header
            fasta_lines = response.text.splitlines()
            sequence = "".join(line.strip() for line in fasta_lines if not line.startswith(">"))
            return sequence
        else:
            return None
    except Exception:
        return None
```

Sequences differ in lenght, so they are padded by adding zeroes.

```python
print(df_no_empty_lines['sequence'].map(len).max())
```

```python
df_no_empty_lines['padded_sequence']=df_no_empty_lines['sequence'].str.ljust(582, '0')
df_padded = df_no_empty_lines
```

Now, the t-scales 5-dimensional vector is assigned to amino acids of every sequence.

```python
t_scales_lookup = pd.read_csv(r'C:\Users\Adrian\Documents\Studium\M1\ai\T_scales_table.csv', skipinitialspace=True)
#t_scales_lookup.head() in case somebody wants to see lookup table
dictionnary = {
    row['symbol']: row[['T_1','T_2','T_3','T_4','T_5']].to_numpy()
    for _, row in t_scales_lookup.iterrows()
}
def encoder(sequence, dictionnary):
    return np.array([dictionnary[char] for char in sequence])
#still for sequence since I'll have it pull PDB sequences once I have enough time to run the PDB sequence lookup
df_padded["encoded_sequence"] = df_padded["padded_sequence"].apply(lambda x: encoder(x, dictionnary))
df_clean = df_padded
df_clean.head()
```

Now, as beforee, a 1-dim CNN is implemented (according to https://www.datacamp.com/tutorial/pytorch-cnn-tutorial, so that noone will complain we don't cite our sources our use chatgpt)

```python
batch = df_clean
print(batch.size)
```

```python
class CNN(nn.Module):
   def __init__(self, in_channels, num_classes):

       """
       Building blocks of convolutional neural network.

       Parameters:
           * in_channels: Number of channels in the input image (for grayscale images, 1)
           * num_classes: Number of classes to predict. In our problem, 10 (i.e digits from  0 to 9).
       """
       super(CNN, self).__init__()

       # 1st convolutional layer
       self.conv1 = nn.Conv2d(in_channels=in_channels, out_channels=8, kernel_size=3, padding=1)
       # Max pooling layer
       self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
       # 2nd convolutional layer
       self.conv2 = nn.Conv2d(in_channels=8, out_channels=16, kernel_size=3, padding=1)
       # Fully connected layer
       self.fc1 = nn.Linear(145, 4728)

   def forward(self, x):
       """
       Define the forward pass of the neural network.

       Parameters:
           x: Input tensor.

       Returns:
           torch.Tensor
               The output tensor after passing through the network.
       """
       x = F.relu(self.conv1(x))  # Apply first convolution and ReLU activation
       x = self.pool(x)           # Apply max pooling
       x = F.relu(self.conv2(x))  # Apply second convolution and ReLU activation
       x = self.pool(x)           # Apply max pooling
       x = x.reshape(x.shape[0], -1)  # Flatten the tensor
       x = self.fc1(x)            # Apply fully connected layer
       x = x.reshape(x.shape[0], -1)  # Flatten the tensor
       x = self.fc1(x)            # Apply fully connected layer
       return x
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"

model = CNN(in_channels=1, num_classes=10).to(device)
print(model)
```

The CNN is trained, using Adam optimizer and MSE as loss function:

```python
X = np.array(df_clean["encoded_sequence"])
y_em = df_clean["emission"].values.astype(np.float32)
print(X.shape)
print(y_em.shape)
```

```python
optimizer = optim.Adam(model.parameters())
loss_fn = nn.MSELoss()
```

```python
num_epochs = 30

for epoch in range(num_epochs):
    print(f"Epoch [{epoch + 1}/{num_epochs}]")
    
    for batch_index, (X_val, y_val) in enumerate(tqdm(zip(batch['encoded_sequence'], batch['emission']))):
        # Convert X_val to a numeric array if it is a list or object
        X_array = np.array(X_val, dtype=np.float32)
        X_tensor = torch.tensor(X_array).unsqueeze(0).to(device)  # add batch dim

        # Convert label
        y_tensor = torch.tensor(y_val, dtype=torch.float32).unsqueeze(0).to(device)

        # Forward pass
        scores = model(X_tensor)
        loss = loss_fn(scores, y_tensor)

        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
```
